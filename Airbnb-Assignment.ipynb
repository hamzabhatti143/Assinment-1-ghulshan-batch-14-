```python
import numpy as np
```

```python
def detect_outliners(col):
    q1= np.percentile(col, 25)
    q3= np.percentile(col, 75)
    iqr = q1 - q3

    lower_boundary = q1 - 1.5 * iqr
    upper_boundary = q3 + 1.5 * iqr

    return lower_boundary, upper_boundary
```

```python
lb, up = detect_outliners(data.host_id)
```

```python
data[(data.host_id < lb) | (data.host_id > up)].drop_duplicates()
```
