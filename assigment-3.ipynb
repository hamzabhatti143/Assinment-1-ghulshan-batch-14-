# Reverse a string using a for loop (no slicing)

    Given Input: "Python"
    
    Expected Output:
    
    Original: Python
    Reversed: nohtyP

```python
user_text = "Python"
reverse = ""

for char in user_text:
    reverse = char + reverse

print(f"Original: {user_text}")
print(f"Reversed: {reverse}")
```

# Count vowels and consonants in a sentence

### Practice Problem: Write a program that counts the total number of vowels and consonants in a given sentence, ignoring spaces and special characters.
    
    Given Input: "Loops are Fun!"
    
    Expected Output:
    
    Vowels: 5
    Consonants: 7

```python
sentence = "Loops are Fun!"

vowels = "aeiouAEIOU"

def text_analyzer():
    vowel = 0
    consonant = 0

    for s in sentence:
        if s.isalpha():   # only letters allowed
            if s in vowels:
                vowel += 1
            else:
                consonant += 1

    print(f"Vowels: {vowel}")
    print(f"Consonants: {consonant}")

text_analyzer()
```

# Count total number of digits in a number
# Practice Problem: Write a program to count the total number of digits in a given integer using a while loop.
    
    Given Input: 75869
    
    Expected Output: Total digits are: 5

```python
num = 75869

count = 0

while num > 0:
    num = num // 10
    count += 1

print("Total digits are:", count)
```

# Find factorial of a number
# Practice Problem: Write a program to use a loop to find the factorial of a given number (e.g., 5!). The factorial of N is the product of all integers from 1 to N.`
    
    Given Input: num = 5
    
    Expected Output: 120

```python
num = 5

factorial = 1

for i in range(1, num + 1):
    factorial *= i

print("Factorial is:", factorial)
```

# Ek function calculator banao jo:

    add
    subtract
    multiply
    divide
    
    ke results print kare.

```python
def calculator(a, b, choice):

    if choice == "add":
        print("Result:", a + b)

    elif choice == "subtract":
        print("Result:", a - b)

    elif choice == "multiply":
        print("Result:", a * b)

    elif choice == "divide":
        if b != 0:
            print("Result:", a / b)
        else:
            print("Cannot divide by zero")

    else:
        print("Invalid choice")

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
choice = input("Choose (add / subtract / multiply / divide): ")

calculator(a, b, choice)
```

# Read File Line by Line Using Loop
# Problem Statement: Write a Python program that reads a file called lines.txt and prints each line one at a time using a loop.
    
    Given Input: A file named lines.txt containing:
    
    Line 1
    Line 2
    Line 3
    Expected Output:
    
    Line 1
    Line 2
    Line 3

```python
file = open("lines.txt", "w")
file.write("""Line 1
Line 2
Line 3""")
with open("lines.txt", "r") as file:
    for line in file:
        print(line.strip())
```

# Append New Sentence to Existing File
# Problem Statement: Write a Python program that appends the sentence This is a new line. to an existing file called notes.txt without overwriting its current content.
    
    Given Input: A file named notes.txt already containing: Original content.
    
    Expected Output: File now contains:
    
    Original content.
    This is a new line.

```python
with open("notes.txt", "a") as file:
    file.write("\nThis is a new line.")

with open("notes.txt", "r") as file:
    for line in file:
        print(line.strip())
```

# Count Total Lines in File
