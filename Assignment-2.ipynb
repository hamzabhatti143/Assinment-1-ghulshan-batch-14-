# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
user_marks = int(input("Enter Your Marks: "))
if user_marks >= 90:
    print("You get A grade")
elif user_marks >= 80:
    print("You get b grade")
elif user_marks >= 70:
    print("You get C grade")
elif user_marks >= 60:
    print("You get D grade")
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day = int(input("Enter day number (1-7): "))

if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
else:
    print("Invalid input")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Enter month number (1-12): "))

if month in (12, 1, 2):
    print("Winter")
elif month in (3, 4, 5):
    print("Spring")
elif month in (6, 7, 8):
    print("Summer")
elif month in (9, 10, 11):
    print("Autumn")
else:
    print("Invalid input")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
num1= int(input("Enter First Number: "))
num2 = int(input("Enter Second Number: "))

operator = input("Please Enter Which Operation you want +, -, *, /: ")
if operator == "+":
    print(f"Your sum of 2 numbers are {num1 + num2}")
elif operator == "-":
    print(f"Your substraction of 2 numbers are {num1 - num2}")
elif operator == "*":
    print(f"Your multiplication of 2 numbers are {num1 * num2}")
elif operator == "/":
    print(f"Your Division of 2 numbers are {num1 / num2}")
else:
    print("Please enter a valid operator which mentioned above.")
```
