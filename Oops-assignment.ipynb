1. Create a class 'BankAccount' with attributes account holder name, account number, and
balance. Add methods for deposit and withdrawal. Implement logic to prevent withdrawal if balance

```python
class BankAccount:
    def __init__(self, acc_name, acc_number, balance=0):
        self.acc_name = acc_name
        self.acc_number = acc_number
        self.balance = balance

    def deposit(self):
        name = input("Enter Your account name: ")
        if name == self.acc_name:
            num = int(input("Enter Account Number: "))
            if num == self.acc_number:
                bal_amount = int(input("Enter Amount which you want to deposit"))
                if bal_amount > self.balance:
                    self.balance += bal_amount
                    print(self.balance)
                else:
                    print("Please Enter positive value")

            else:
                print("Please Enter Correct Account Number")
        else:
            print("Your Account Name is not Exist in Our System")

    def withdraw(self):
        name = input("Enter Your account name: ")
        if name == self.acc_name:
            num = int(input("Enter Account Number: "))
            if num == self.acc_number:
                bal_amount = int(input("Enter Amount which you want to withdraw"))
                if bal_amount <= self.balance and bal_amount >= 500:
                    self.balance += bal_amount
                    print(self.balance)
                else:
                    print(f"You need deposit first. your balance is {self.balance}")

            else:
                print("Please Enter Correct Account Number")
        else:
            print("Your Account Name is not Exist in Our System")

ba = BankAccount(acc_name="Hamza", acc_number=1234)
user_opt = input("Which Task You need to perfrom 1: Deposit 2: WithDraw")
if user_opt == "deposit" or user_opt == "Deposit":
    ba.deposit()
else:
    ba.withdraw()
```

6. Design an abstract class 'Shape' with a method 'area()'. Create subclasses 'Circle' and
'Rectangle' and implement area calculation logic.

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def shap(self):
        print("The Shape of the circle is round")

    def area(self):
        return 3.14 * self.radius * self.radius
        
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
        
    def area(self):
        return self.length * self.width

circle = Circle(5)
print("Area of Circle:", circle.area())

rectangle = Rectangle(10, 5)
print("Area of Rectangle:", rectangle.area())
```

8. Build a class 'ShoppingCart' where user can add items with price. Implement logic to calculate
total bill and apply discount if total exceeds a certain amount.

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item, price):
        self.items.append((item, price))

    def calculate_total(self):
        total = 0

        for item, price in self.items:
            total += price

        if total > 100:
            discount = total * 0.10    # 10% discount
            total -= discount
            print(f"Discount Applied: ${discount}")

        print(f"Total Bill: ${total}")


cart = ShoppingCart()

cart.add_item("Shoes", 50)
cart.add_item("T-Shirt", 30)
cart.add_item("Watch", 40)

cart.calculate_total()
```

10. Design a class 'Library' where users can borrow and return books. Add logic to check
availability before issuing a book.

```python
class Library:

    def __init__(self):
        self.books = {}

    
    def add_book(self, book_name, quantity):
        if book_name in self.books:
            self.books[book_name] += quantity
        else:
            self.books[book_name] = quantity

    
    def is_available(self, book_name):
        return self.books.get(book_name, 0) > 0

    
    def borrow_book(self, book_name):
        if self.is_available(book_name):
            self.books[book_name] -= 1
            print(f"You borrowed '{book_name}' ")
        else:
            print(f"'{book_name}' is not available ")

    def return_book(self, book_name):
        if book_name in self.books:
            self.books[book_name] += 1
        else:
            self.books[book_name] = 1
        print(f"You returned '{book_name}' ")

    def show_books(self):
        print("\nLibrary Books:")
        for book, qty in self.books.items():
            print(f"{book} : {qty}")

library = Library()

library.add_book("Python Basics", 3)
library.add_book("OOP Concepts", 2)

library.show_books()

library.borrow_book("Python Basics")
library.borrow_book("OOP Concepts")

library.borrow_book("Data Science")

library.return_book("Python Basics")

library.show_books()
```

```python
class Bank:
    def __init__(self, bank_name):
        self.bank_name = bank_name
        self.branches = []

    def add_branch(self, branch):
        if branch not in self.branches:
            self.branches.append(branch)
            print(f"{branch} branch added successfully.")
        else:
            print(f"{branch} branch already exists.")

    def remove_branch(self, branch):
        if branch in self.branches:
            self.branches.remove(branch)
            print(f"{branch} branch removed successfully.")
        else:
            print(f"{branch} branch not found.")

    def display_branches(self):
        print(f"\n{self.bank_name} Branches:")
        if self.branches:
            for branch in self.branches:
                print("-", branch)
        else:
            print("No branches available.")



bank = Bank("HBL Bank")


bank.add_branch("Karachi")
bank.add_branch("Lahore")
bank.add_branch("Islamabad")


bank.display_branches()


bank.remove_branch("Lahore")


bank.display_branches()
```

```python
from abc import ABC, abstractmethod
import math


class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

class Triangle(Shape):

    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

circle = Circle(5)
triangle = Triangle(10, 6)

print("Area of Circle:", circle.area())
print("Area of Triangle:", triangle.area())
```

```python
from abc import ABC, abstractmethod
import uuid

class Product(ABC):

    def __init__(self, name, price, stock):
        self.name = name
        self.__price = price      # encapsulation
        self.stock = stock

    def get_price(self):
        return self.__price

    def reduce_stock(self, qty):
        self.stock -= qty

    def check_stock(self, qty):
        return self.stock >= qty

    def apply_tax(self):
        return self.__price * 1.18  # 18% tax

    @abstractmethod
    def get_discount(self):
        pass

class AC(Product):
    def get_discount(self):
        return 0.10  # 10%


class Refrigerator(Product):
    def get_discount(self):
        return 0.08  # 8%


class WashingMachine(Product):
    def get_discount(self):
        return 0.05  # 5%

class Customer:

    def __init__(self, name):
        self.name = name
        self.cart = []
        self.invoice_id = str(uuid.uuid4())

    def add_to_cart(self, product, qty):
        if product.check_stock(qty):
            self.cart.append((product, qty))
            product.reduce_stock(qty)
        else:
            print(f"{product.name} out of stock!")

    def generate_invoice(self):
        total = 0

        print("\n------ INVOICE ------")
        print("Customer:", self.name)
        print("Invoice ID:", self.invoice_id)

        for product, qty in self.cart:
            price = product.get_price()

            discount = price * product.get_discount()
            final_price = (price - discount) * qty
            final_price = final_price * 1.18  # tax

            total += final_price

            print(f"{product.name} x {qty} = {final_price}")

        print("TOTAL BILL:", total)

        self.save_invoice(total)

    def save_invoice(self, total):
        with open("invoices.txt", "w") as f:
            f.write(f"""
            Customer: {self.name}
            Invoice ID: {self.invoice_id}
            Total Bill: {total}
            ------------------------""")

ac = AC("Haier AC", 50000, 5)
fridge = Refrigerator("Dawlance Fridge", 40000, 3)
wm = WashingMachine("LG Washer", 30000, 2)

customer = Customer("Hamza")

customer.add_to_cart(ac, 1)
customer.add_to_cart(fridge, 1)
customer.add_to_cart(wm, 1)

customer.generate_invoice()
```

```python
from abc import ABC, abstractmethod
import uuid


class FoodItem(ABC):

    def __init__(self, name, price, stock):
        self.name = name
        self.__price = price
        self.__stock = stock

    def get_price(self):
        return self.__price

    def check_stock(self, qty):
        return self.__stock >= qty

    def reduce_stock(self, qty):
        self.__stock -= qty

    def get_total(self, qty):
        total = self.__price * qty

        if total > 2000:
            total -= total * 0.10  # 10% discount

        return total * 1.15  # 15% tax

    @abstractmethod
    def prepare(self):
        pass

class Pizza(FoodItem):
    def prepare(self):
        print("Pizza is being baked")


class Burger(FoodItem):
    def prepare(self):
        print("Burger is being grill")


class Drink(FoodItem):
    def prepare(self):
        print("Drink is being prepared")
class Customer:

    def __init__(self, name):
        self.name = name
        self.cart = []
        self.order_id = str(uuid.uuid4())

    def add_to_cart(self, item, qty):
        if item.check_stock(qty):
            self.cart.append((item, qty))
            item.reduce_stock(qty)
        else:
            print(f"{item.name} is out of stock!")

    def checkout(self, method="cash"):
        total = 0

        print("\n------ INVOICE ------")
        print("Customer:", self.name)
        print("Order ID:", self.order_id)
        print("Payment Method:", method)

        for item, qty in self.cart:
            item.prepare()
            bill = item.get_total(qty)
            total += bill
            print(f"{item.name} x {qty} = {bill}")

        print("TOTAL BILL:", total)

        self.save_order(total)
        
    def save_order(self, total):
        with open("orders.txt", "a") as f:
            f.write(f"""
            Customer: {self.name}
            Order ID: {self.order_id}
            Total: {total}
            ------------------------""")

pizza = Pizza("Pepperoni Pizza", 1200, 5)
burger = Burger("Zinger Burger", 800, 10)
drink = Drink("Cola", 200, 20)

customer = Customer("Hamza")

customer.add_to_cart(pizza, 2)
customer.add_to_cart(burger, 1)
customer.add_to_cart(drink, 3)

customer.checkout("card")
```
